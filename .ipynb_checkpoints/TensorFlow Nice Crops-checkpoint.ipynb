```python
def create_graph():
    with gfile.FastGFile(os.path.join(model_dir, 'classify_image_graph_def.pb'), 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
        _ = tf.import_graph_def(graph_def, name='')
create_graph()
```

```python
def extract_features(list_images):
    nb_features = 2048
    features = []
    labels = []

    next_to_last_tensor = sess.graph.get_tensor_by_name('pool_3:0')

    for ind, image in enumerate(list_images):
        if ind in to_ignore:
            mini_features = []
            for i in range(2):
                mini_features.append(np.zeros(nb_features))
            features.append(mini_features[:])
            for (i,x) in enumerate(features[-1]):
                print(i,x)
            continue
        if (ind%1 == 0):
            print('Processing %s...' % (image))
        if not gfile.Exists(image):
            tf.logging.fatal('File does not exist %s', image)

        image_data = gfile.FastGFile(image, 'rb').read()
        image_data_d = tf.image.decode_jpeg(image_data)
        
        of_y = int(sess.run(image_data_d).shape[0]/4)
        of_x = int(sess.run(image_data_d).shape[1]/5)
        
        mini_features = []
        
        for i in range(4):
            if i == 0 or i == 3:
                continue
            for j in range(5):
                if j == 0 or j == 4 or j == 1 or j == 3:
                    continue
        
                image_data_de  = tf.image.crop_to_bounding_box(image_data_d,i*of_y,j*of_x,of_y,of_x)
                #print(sess.run(image_data_de).shape[0],sess.run(image_data_de).shape[1])
                #image_data_c = tf.image.encode_jpeg(image_data_de)
                image_data_c = tf.image.encode_jpeg(image_data_de)
                predictions = sess.run(next_to_last_tensor,{'DecodeJpeg/contents:0': sess.run(image_data_c)})
                #display.display(Image.open(io.BytesIO(sess.run(image_data_c))))
                #print(predictions)
                mini_features.append(np.squeeze(predictions))
        
        features.append(mini_features[:])
        for (i,x) in enumerate(features[-1]):
            print(i,x)
        labels.append(io.BytesIO(image_data))
                #labels.append(re.split('_\d+',image.split('/')[1])[0])
    return features,labels
```

```python
query_dir = 'db/robes/mod/'
list_query = [query_dir+f for f in os.listdir(query_dir) if re.search('jpg|JPG', f)]
list_query = list(filter(lambda x: "crop" in x,list_query))
list_query.sort(key=lambda x: int(x.split("/")[-1].split("_")[0]))
print(len(list_query))
```

```python
features_cat = extract_features(list_images)
```

```python
features_query = extract_features(list_query)
```

```python
print(len(features_query[0]),len(features_cat[0]))
```

```python
def sim(vecA,vecB):
    return vecA.dot(vecB)
```

```python
def query(i_query,p=False):
    #if i_query in to_ignore:
        #return False
    sim_vec = []
    for i in range(len(features_cat[0])):
        if i in to_ignore:
            sim_vec.append(0.0)
            continue
        M = 0.0
        for k in range(2):
            for l in range(2):
                M = max(M,sim(features_query[0][i_query][k],features_cat[0][i][l]))
        sim_vec.append(M)
    sim_vec = np.array(sim_vec)
    arg_s = sim_vec.argsort()[:-10:-1]
    if p:
        print(i_query,arg_s,[sim_vec[i] for i in arg_s], sim_vec[i_query])
    return i_query in arg_s
```

```python
for i in range(len(features_cat[0])):
    if not i in to_ignore:
        for j in range(2):
            features_cat[0][i][j] /= np.linalg.norm(features_cat[0][i][j])
            features_query[0][i][j] /= np.linalg.norm(features_query[0][i][j])

res = 0
k = 0
for i in range(len(features_cat[0])):
    if not i in to_ignore:
        res += query(i)
    else:
        k += 1
res /= (len(features_cat[0])-k)
print(res)
```
